```python
import qrcode

qr = qrcode.QRCode(
    version = 15, #higher the number, higher the code image and complication
    box_size = 10,
    border = 5
)

data = "https://www.youtube.com/watch?v=onHPipeASdk&list=PLpp8-k7G_6Y3Wj1suZQ-9lATFzFuGw93x"
qr.add_data(data)
qr.make(fit = True)
img = qr.make_image(fill = "black", back_color = "white")
img.save("test.png")
```
